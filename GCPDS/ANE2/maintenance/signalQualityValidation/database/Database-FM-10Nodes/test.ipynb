```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
# ═══════════════════════════════════════════════════════════════════════
#  READ ALL CSV FILES FROM db/ DIRECTORY
# ═══════════════════════════════════════════════════════════════════════

# Path to db directory (relative to this notebook)
db_dir = "data"

print("=" * 80)
print("📂 READING CSV FILES FROM db/ DIRECTORY")
print("=" * 80)

# Verify path exists
if not os.path.exists(db_dir):
    print(f"❌ Directory not found: {db_dir}")
    print(f"   Absolute path would be: {os.path.abspath(db_dir)}")
else:
    print(f"✓ Directory found: {db_dir}")
    print(f"  Full path: {os.path.abspath(db_dir)}\n")

# List all CSV files in the directory
csv_files = [f for f in os.listdir(db_dir) if f.endswith('.csv')]
csv_files = sorted(csv_files)

print(f"📋 Found {len(csv_files)} CSV files:")
print("-" * 80)
for i, fname in enumerate(csv_files, 1):
    filepath = os.path.join(db_dir, fname)
    file_size = os.path.getsize(filepath) / (1024 * 1024)  # Size in MB
    print(f"  {i}. {fname:<40} ({file_size:.2f} MB)")

# Read all CSV files into a dictionary
print(f"\n📥 Loading all CSV files...")
print("-" * 80)

csv_data = {}
for fname in csv_files:
    filepath = os.path.join(db_dir, fname)
    try:
        df = pd.read_csv(filepath, encoding='utf-8-sig')
        csv_data[fname] = df
        print(f"  ✓ {fname:<40} → {df.shape[0]:>7} rows × {df.shape[1]:>3} cols")
    except Exception as e:
        print(f"  ✗ {fname:<40} → ERROR: {e}")

print("-" * 80)
print(f"\n✓ Successfully loaded {len(csv_data)} files\n")

# Display summary of first file
if csv_data:
    first_file = list(csv_data.keys())[0]
    first_df = csv_data[first_file]
    print(f"📊 Sample data from '{first_file}':")
    print(f"   Columns: {list(first_df.columns)}")
    print(f"   Shape: {first_df.shape}")
    print(f"   Data types:\n{first_df.dtypes}\n")
    print(f"   First row:")
    print(first_df.head(1).to_string())
```
